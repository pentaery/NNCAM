```python
from pathlib import Path
import xarray as xr

data_path = Path("/home/ET/mnwong/ML/data/Qobs10_SPCAMM.000.cam.h1.0001-02-13-00800.nc")
with xr.open_dataset(data_path) as ds:
    display(ds)
    print("\nData variables:", list(ds.data_vars))
    print("Coordinates:", list(ds.coords))
```

```python
import numpy as np
U = np.array(ds['CLDLIQ'])
print(U.max(), U.min(), U.mean(), U.std())
import matplotlib.pyplot as plt
hist, bins = np.histogram(U, bins=10000, range=(U.min(), U.max()))
plt.bar(bins[:-1], hist, width=np.diff(bins), edgecolor='black')
plt.xlabel('U')
plt.ylabel('Frequency')
plt.title('Histogram of U')
plt.show()
```
